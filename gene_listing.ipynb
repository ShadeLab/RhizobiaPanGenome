```python
import os
import numpy as np

def groupScaffolds(rootdir):
    store = []
    curspec = ""
    curscaffs = []
    startcheck = 0
    count = 0
    for subdir, dirs, files in os.walk(rootdir):
        print(count)
        count = count + 1
        if(startcheck == 0):
            curspec = str(subdir)
            curscaffs.append(curspec)
            startcheck = 1
        elif(curspec != str(subdir)):
            curspec = str(subdir)
            store.append(curscaffs)
            curscaffs = []
            curscaffs.append(curspec)
        for i in range(len(files)):
            if i == 0:
                tempO = open(os.path.join(subdir, files[i]))
                line = tempO.readline()
                while line:
                    linecheck = 0 
                    curname = ""
                    for i in range(len(line)):
                        if(line[0] == '>'):
                            linecheck = 1
                        if(linecheck == 1 and i != 0):
                            curname = curname + line[i]
                    if(linecheck == 1):
                        curscaffs.append(curname)
                    line = tempO.readline()     
    store.append(curscaffs)
                            
    return(store)
```

```python
storelist2 = groupScaffolds('RJGI')
```

```python
import os
import numpy as np

def geneMatch(blastnfasta,scaftable,savelist,namelist,genecol):
    
    #out only matches convert to fasta file from blastn checked for matches
    
    blafas = open(blastnfasta)
    
    line = blafas.readline()
    
    count = 0
    
    while(line):
        if(line[0] == ">"):
            curtag = ""
            for i in range(1,len(line)-2):   #pos 1 is '>' last two positions are a space and '\n' 
                curtag = curtag + line[i]
            for i in range(len(scaftable)):
                curlab = scaftable[i] 
                for j in range(1,len(curlab)):
                    if(curtag == curlab[j]):
                        count = count + 1
                        currow = savelist[i]
                        currow[genecol] = currow[genecol] + 1
                        savelist[i] = currow     
                        curnamerow = namelist[i]
                        curnamelist = curnamerow[genecol]
                        curnamelist.append(curtag)
                        curnamerow[genecol] = curnamelist
                        namelist[i] = curnamerow
        line = blafas.readline()
        
    print(count)
    return(savelist,namelist)
```

```python
#This function will remove and genes from fastas that do not have the other 2 genes present. It will also write the genome name
#to the beginning of the tag. This will make aligning and merging a lot easier. 

import numpy as np
import os

def genePresOnly(infasta,outfasta,NT3,genecol):
    
    #if genecol=1: 16srRNA, 2: recA, 3: glnII
    
    infas = open(infasta)
    outfas = open(outfasta,'w')
    
    line = infas.readline()
    while(line):
        mathcheck = 1
        if(line[0] == '>'):
            curtag = ""
            for i in range(1,len(line)-2):
                curtag = curtag + line[i]
            for i in range(1,(len(NT3))):
                currow = NT3[i]
                curnamelist = currow[genecol]
                for j in range(len(curnamelist)):
                    curname = curnamelist[j]
                    if(curname == curtag):
                        outfas.write('>')
                        outfas.write(currow[0])
                        outfas.write(': ')
                        outfas.write(curtag)
                        outfas.write('\n')
                        line = infas.readline()
                        outfas.write(line)
                        matchcheck = 0
        if(mathcheck):
            line = infas.readline()
    infas.close()
```
